```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_excel("df.xlsx")

# --- Step 1: RFM Scoring ---
# Recency: lower days = better
df['R_Score'] = pd.qcut(df['RecencyDays'], 5, labels=[5,4,3,2,1])

# Frequency: constant, so assign 1
df['F_Score'] = 1

# Monetary: higher spend = better
df['M_Score'] = pd.qcut(df['Monetary'], 5, labels=[1,2,3,4,5])

# --- Step 2: Combined RFM Segment ---
df['RFM_Concat'] = (
    df['R_Score'].astype(str) +
    df['F_Score'].astype(str) +
    df['M_Score'].astype(str)
)

# --- Step 3: Business Rule Mapping ---
segment_map = {
    "555": "Champion",
    "515": "High Spender",
    "511": "Potential Loyalist",
    "151": "At Risk",
    "111": "Lost",
    # extend rules as needed
}
df['Segment'] = df['RFM_Concat'].map(segment_map).fillna("Other")

# --- Step 4: Segment Counts ---
segment_counts = df['Segment'].value_counts()
print("Customer Count per Segment:\n", segment_counts)

# --- Step 5: Revenue Contribution per Segment ---
segment_revenue = df.groupby('Segment')['Monetary'].sum().sort_values(ascending=False)
print("\nRevenue Contribution per Segment:\n", segment_revenue)

# --- Step 6: Scatter Plot (Recency vs Monetary) ---
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='RecencyDays', y='Monetary', hue='Segment', palette='Set2')
plt.title("Recency vs Monetary by Segment")
plt.show()

# --- Step 7: Pareto Analysis ---
df_sorted = df.sort_values(by="Monetary", ascending=False).reset_index(drop=True)
df_sorted["CumRevenue"] = df_sorted["Monetary"].cumsum()
df_sorted["CumRevenuePerc"] = df_sorted["CumRevenue"] / df_sorted["Monetary"].sum() * 100

fig, ax1 = plt.subplots(figsize=(12,6))
ax1.bar(df_sorted.index, df_sorted["Monetary"], color="steelblue")
ax1.set_ylabel("Revenue per Customer", color="steelblue")
ax1.set_xlabel("Customers (sorted by spend)")

ax2 = ax1.twinx()
ax2.plot(df_sorted.index, df_sorted["CumRevenuePerc"], color="orange", marker="o")
ax2.axhline(80, color="green", linestyle="--", label="80% Threshold")
ax2.set_ylabel("Cumulative % of Revenue", color="orange")

plt.title("Pareto Chart - Customer Revenue Contribution")
ax2.legend(loc="best")
plt.show()

```
